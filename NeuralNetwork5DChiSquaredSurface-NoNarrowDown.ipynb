```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker, cm
import seaborn as sns
from keras.layers import Input, Dense
from keras.models import Model
import corner

#Load data and seperate the input and output data

data = np.loadtxt('5DchiSqSurfaceTrainingDataUnrestricted.dat')
inputData = data[:, 0:5]
outputData = data[:, 5]

plt.figure(figsize=(6, 6))

#Needs a corner plot showing the variables

labels = ["Av", "As", "Ac", "Aa", "Ap"]
corner.corner(inputData, labels=labels)

print("This is the training data being given to the neural network")
```

```python
#Set up neural network structure

epochs = 200
batch_size = 16
input_dim = 5

input_layer = Input(shape=(input_dim))
layer1 = Dense(30, activation="relu", kernel_initializer="glorot_uniform")(input_layer)
layer2 = Dense(25, activation="relu", kernel_initializer="glorot_uniform")(layer1)
output_layer = Dense(1, activation="linear", kernel_initializer="glorot_uniform")(layer2)
                                                                                
nn = Model(inputs=input_layer, outputs=output_layer)
nn.compile(optimizer="adam", loss="mean_squared_error", metrics = ["mse"])

#train neural network

history = nn.fit(
    inputData,
    outputData,
    epochs=epochs,
    batch_size=batch_size,
    shuffle=True,
    verbose=1,
    validation_split= 0.20,
).history

#Plot the MSE against the epochs

train_loss = history["loss"]
val_loss = history["val_loss"]
t = np.linspace(1, len(train_loss), len(train_loss))

plt.figure(figsize=(8, 8))
sns.lineplot(x = t, y = train_loss, label="train")
sns.lineplot(x = t, y = val_loss, label="validation")
plt.xlabel("Epochs")
plt.title("MSE")
plt.legend() 
plt.show()

#print the final MSE for the full set of training data

print(np.mean((outputData - nn.predict(inputData)[:,0])**2))
print(nn.summary())
```

```python
limits = [[  0, 50],
          [-50,  0],
          [-50,  0],
          [-50,  0],
          [-50,  0]]
```

```python
import emcee
import math

def log_prior(theta, limits):
    Av, As, Ac, Aa, Ap = theta
    if limits[0][0] <= Av <= limits[0][1] and limits[1][0] <= As <= limits[1][1] and limits[2][0] <= Ac <= limits[2][1] and limits[3][0] <= Aa <= limits[3][1] and limits[4][0] <= Ap <= limits[4][1]:
        return 0.0
    return -np.inf
    
def log_prob(p0, limits):
    lp = log_prior(p0, limits)
    if not np.isfinite(lp):
        return -np.inf
    return lp - math.exp(nn.predict(p0.reshape(1, 5)))
```

```python
nwalkers = 25
ndim = 5
p0 = np.random.rand(nwalkers, ndim)
p0[:, 0] = p0[:, 0]*(limits[0][1] - limits[0][0]) + limits[0][0]
p0[:, 1] = p0[:, 1]*(limits[1][1] - limits[1][0]) + limits[1][0]
p0[:, 2] = p0[:, 2]*(limits[2][1] - limits[2][0]) + limits[2][0]
p0[:, 3] = p0[:, 3]*(limits[3][1] - limits[3][0]) + limits[3][0]
p0[:, 4] = p0[:, 4]*(limits[4][1] - limits[4][0]) + limits[4][0]
```

```python
import time

sampler = emcee.EnsembleSampler(nwalkers, ndim, log_prob, args=[limits])

start = time.perf_counter()
sampler.run_mcmc(p0, 250, progress=True)
stop = time.perf_counter()

print("Done")
stop = time.perf_counter()
print("Time taken: ", stop - start)
```

```python
flat_samples = sampler.get_chain(flat=True, discard=50)
print(np.shape(flat_samples))
labels = ["Av", "As", "Ac", "Aa", "Ap"]
fig = corner.corner(flat_samples, labels=labels, quantiles=[0.16, 0.5, 0.84], show_titles=True)
```

```python
def modeOfData(flat_samples):
    
    peaks = []
    
    a = np.histogram(flat_samples, bins=10)
    totalCounts = np.sum(a[0])
    
    while True:
        
        maxIdx = np.where(a[0] == a[0].max())[0][0]
        
        if a[0][maxIdx] / totalCounts >= 0.05:
            
            aLeft = a[1][maxIdx]
            aRight = a[1][maxIdx + 1]
            aMid = ((aRight - aLeft) / 2) + aLeft
            peaks.append(aMid)

            a[0][maxIdx] = 0
            
        else:
            break

    return peaks
```

```python
measuredNuclearData = np.loadtxt("ZNBE.dat")

for i in measuredNuclearData:
    
    i[2] = i[2] * (i[0] + i[1])

AvPeaks = modeOfData(flat_samples[:, 0])
AsPeaks = modeOfData(flat_samples[:, 1])
AcPeaks = modeOfData(flat_samples[:, 2])                           
AaPeaks = modeOfData(flat_samples[:, 3])
ApPeaks = modeOfData(flat_samples[:, 4])

SEMFData = []
chiSq = 0
chiSqSurfaceData = []

for Av in AvPeaks:
    for As in AsPeaks:
        for Ac in AcPeaks:
            for Aa in AaPeaks:
                for Ap in ApPeaks:
                    
                    chiSq = 0
                    SEMFData = []
                
                    for i in measuredNuclearData:

                        if i[0] + i[1] < 16:
                            continue

                        Z = i[0]
                        N = i[1]
                        A = N + Z

                        vol = Av * A
                        sur = As * A**(2/3)
                        cou = Ac * (Z * (Z - 1)) * A**(-1/3)
                        asy = Aa * (N - Z)**2 * A**(-1)

                        if Z%2==0 and N%2==0:
                            pai = Ap * A**-0.5
                        elif Z%2!=0 and N%2!=0:
                            pai = -Ap * A**-0.5
                        else:
                            pai = 0

                        BE = vol + sur + cou + asy + pai
                        SEMFData.append([A, BE])
                        chiSq += (BE - i[2])**2
                    
                    chiSqSurfaceData.append([Av, As, Ac, Aa, Ap, chiSq, SEMFData])
                
plt.figure(figsize=(12,12))

chiSq = [i[5] for i in chiSqSurfaceData]
minIdx = chiSq.index(min(chiSq))
SEMFData = chiSqSurfaceData[minIdx][6]

A = [i[0] for i in SEMFData]
BE = [i[1]/i[0] for i in SEMFData]

plt.scatter(A, BE, s=5, label='Coefficient Predictions')

realData = [([i[0]+i[1],i[2]]) for i in measuredNuclearData if (i[0] + i[1] >= 16)]
A = [i[0] for i in realData]
BE = [i[1]/i[0] for i in realData]

plt.scatter(A, BE, s = 5, label='Measured Nuclear Data')
plt.legend()
plt.show()

print(chiSqSurfaceData[minIdx][0:5])
```
